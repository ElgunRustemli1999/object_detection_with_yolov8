```python
# tek bir img detection etmek ucun
import cv2
import imutils
from ultralytics import YOLO

font = cv2.FONT_HERSHEY_SIMPLEX
img_path = r'C:\Users\Gold\Desktop\yolov8_gpu\1_object_detection\inferance\88.jpg'
model_path= r'C:\Users\Gold\Desktop\yolov8_gpu\1_object_detection\runs\detect\yolov8_brain_tumors_detection3\weights\best.pt'

model = YOLO(model_path)

img = cv2.imread(img_path)
img=imutils.resize(img,width=360)

results  = model(img)[0]

treshold = 0.5

for result in results.boxes.data.tolist():
    x1,y1,x2,y2,score,class_id = result
    x1,y1,x2,y2,class_id = int(x1),int(y1),int(x2),int(y2),int(class_id)
    if score>treshold:
        cv2.rectangle(img,(x1,y1),(x2,y2),(0,255,0),2)

        class_name = results.names[class_id]
        score = score*100
        text = f'{class_name}: %{score:.2f}'
        cv2.putText(img,text,(x1,y1-10),font,0.5,(0,255,0),1,cv2.LINE_AA)
cv2.imshow("img",img)
```
